```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
image = cv2.imread('images.jpg')


'''
opencv默認的imread是以BGR的方式進行存儲的,
而matplotlib的imshow默認則是以RGB格式展示,
所以此處我們必須對圖片的通道進行轉換。
'''
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
height, width = image.shape[:2]
```

```python
height
```

```python
width
```

```python
#Rotation
rotation_Matrix = cv2.getRotationMatrix2D((width/2, height/2), 30, 0.6)
rotated_img = cv2.warpAffine(image, rotation_Matrix, (width, height))

plt.imshow(rotated_img)
plt.title('Rotated image')
plt.show()
```
